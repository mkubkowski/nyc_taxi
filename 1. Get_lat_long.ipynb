```python
import geopandas as gpd
import pandas as pd
```

```python
borough = gpd.read_file('G:/dane_python/nyc_taxi_other/taxi_zones/taxi_zones.shp')
borough = borough.to_crs(epsg=4326)
```

```python
borough
```

```python
borough['longitude'] = borough.centroid.x
borough['latitude'] = borough.centroid.y
```

```python
type(borough)
```

```python
df = pd.DataFrame(borough.loc[:,['OBJECTID','Shape_Leng','Shape_Area','zone','LocationID','borough','longitude','latitude']])
```
